```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Salary_dt.csv")
data.head()
```

```python
ind_var = data.drop("salary_more_then_100k", axis=1)
tar_var = data.salary_more_then_100k
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
ind_var["company_le"] = le_company.fit_transform(ind_var["company"])
ind_var["job_le"] = le_job.fit_transform(ind_var["job"])
ind_var["degree_le"] = le_degree.fit_transform(ind_var["degree"])
ind_var.head()
```

```python
ind_var_num = ind_var.drop(["company","job","degree"], axis=1)
ind_var_num
```

# Training Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(ind_var_num,tar_var, test_size = 0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.score(X_test, y_test)*100
```

```python
model.predict([[2,2,1]])
```

# Using Dummy Variables

```python
ind_var = data.drop("salary_more_then_100k", axis=1)
tar_var = data.salary_more_then_100k
```

```python
ind_var
```

```python
ind_var_com_dummies = pd.get_dummies(ind_var.company, prefix ="company", drop_first = True)
ind_var_job_dummies = pd.get_dummies(ind_var.job, prefix ="company", drop_first = True)
ind_var_degree_dummies = pd.get_dummies(ind_var.degree, prefix ="company")
```

```python
new_df = pd.concat([ind_var,ind_var_com_dummies,ind_var_job_dummies,ind_var_degree_dummies], axis=1)
new_df
```

```python
final = new_df.drop(["company", "job","degree"], axis=1)
final
```

```python
X_train, X_test, y_train, y_test = train_test_split(final,tar_var, test_size = 0.30)
```

```python
model2 = DecisionTreeClassifier()
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
y_pred
```

```python
y_test
```

```python
model2.score(X_test, y_test)*100
```
